```r
#2

# Obliczenie mapy odległości każdego punktu w siatce od najbliższego uskoku
distance_map <- distmap(v.faults)

plot(distance_map, main = "Mapa Z(u): Odległość od najbliższego uskoku")
plot(v.gold, add = TRUE, col = 'gold')
```

```r
Z_u <- distfun(v.faults)


rho_estimation <- rhohat(v.gold, Z_u)


plot(rho_estimation, 
     xlab = "Odległość od najbliższego uskoku", ylab = "Intensywność", col = "blue", lwd = 2)
```

```r
distance_area <- distfun(v.faults)

plot(distance_area, main = "Odległość od najbliźszego uskoku")
plot(v.gold, add = TRUE, col = 'gold')
```
